# Correlation Illusion

This shows how cherry picking specific data subsets can hide strong negative correlations.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data into data frame
df = pd.read_csv(r"..\datasets\cereal.csv")
```

Let us create yet another hypothetical situation. 

Suppose a company wants to argue that high-sugar cereals are actually better received by consumers.

Although the full dataset shows a negative correlation between sugar content and rating, the company could selectively subset the data. For example, focusing only on cereals above a certain sugar threshold, and highlight patterns within that restricted group.

By presenting only this filtered view, they could construct a narrative suggesting that higher sugar content is associated with higher ratings, while omitting the broader context.

```python
# susbet data to contain cereals with high sugar
df_sugar = df[df['sugars'] > 6]
```

```python
# create a deceptive scatter plot
plt.scatter(x=df_sugar['fiber'], y=df_sugar['rating'])

# label axis
plt.xlabel('Fiber content [grams]')
plt.ylabel('Rating')

# add title 
plt.title('High Sugar Subset: Fiber = Higher Rating!')

plt.show()
```

By selectively choosing a subset of the data, we can shift attention from one variable to another and construct a different story, without falsifying any data.

```python
# Plotting the 'Truth' vs the 'Illusion'
plt.scatter(df['fiber'], df['rating'], color='orange', label='Rest of Market')
plt.scatter(df_sugar['fiber'], df_sugar['rating'], color='blue', label='Our High Fiber Sugar Bombs')

plt.title("Full Market View")
plt.legend()
plt.show()
```

When viewed together, it becomes clear that even the best high sugar cereals are still underperforming compared to most low-sugar cereals.



Now to the honest scatter plot.

```python
plt.scatter(x=df['sugars'], y=df['rating'])

# label axis
plt.xlabel('Sugar content [grams]')
plt.ylabel('Rating')

# add title 
plt.title('Honest Scatter Plot')

plt.show()
```

In the complete dataset, there is a strong negative correlation between sugar and ratings. Generally, as sugar content increases, the rating drops significantly.
