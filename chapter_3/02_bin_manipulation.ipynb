# Bin Manipulation

This shows how manipulating the width of histogram bins affect how we view data.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset into data frame
df = pd.read_csv('cereal.csv')

df.head()
```

Let us create a hypothetical situation.

A competitor brand is trying to show that they are better than **Kellogs**. They try to compare the sugar content in their cereals to that of **Kellogs**.

They decide to use a histogram to show the distribution.

```python
# assign cereals made by kellogs to a single variable
df_kellogs = df[df['mfr'] == 'K']
```

```python
# create misleading histogram
plt.hist(df_kellogs['sugars'], bins=5)

# label all axis
plt.xlabel('Amount of sugar (grams)')
plt.ylabel('Count')

# add title
plt.title('Deceptive Chart: 5 bins ')

plt.show()
```

In this histogram, the number of bins are reduced (5 bins), combining several sugar ranges into a single category.

This makes the distribution appear heavily concentrated in higher sugar ranges, exaggerating the perception that most cereals made by **Kellogs** are highly sweetened.

```python
# create honest histogram
plt.hist(df_kellogs['sugars'], bins=10)

# label all axis
plt.xlabel('Amount of sugar (grams)')
plt.ylabel('Count')

# add title
plt.title('Honest Chart: 10 bins ')

plt.show()
```

Unlike the competitor’s chart, which had fewer bins and made it look like all their cereals are high in sugar, this version uses 10 bins and reveals that there are two clear peaks. 

The first peak, around 6–7.5 grams, likely represents healthier cereals aimed at adults, while the second peak, around 12–13.5 grams, likely corresponds to kids’ cereals.

By increasing the number of bins, you can see a higher resolution view. Cereals made by **Kellogs** are not just average in sugar, they are either fairly low sugar or quite sugary. 

This shows why it’s dangerous to trust a chart without looking at how the data is being presented.
