## Truncated Axis 

This demonstrates how manipulating the Y-axis range can exaggerate small trends.

```python
# import libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# load data into dataframe
df = pd.read_csv("cereal.csv")
df.head()
```

Let us create a hypothetical situation.

Let's say **Smacks** was running a campaign and are trying to prove that their cereal contains more potassium than most cereal brands.

They compare their potassium content to that of **Total Corn Flakes**.



```python
# assign cereal brands to data frame
df_cereals = df.iloc[[66, 69]]
```

```python
# create deceptive bar chart
plt.bar(x =df_cereals['name'],  height=df_cereals['potass'])

# add misleading y-axis 
plt.ylim(34, 40)

# add axis labels
plt.xlabel("Cereal")
plt.ylabel('Amount of Potassium (mg)')

# add title
plt.title('Misleading Bar Chart')

plt.show()
```

Changing the limits of the y-axis makes such a small difference in potassium content look very huge

```python
# honest bar chart
plt.bar(x =df_cereals['name'],  height=df_cereals['potass'])

# add axis labels
plt.xlabel("Cereal")
plt.ylabel('Amount of Potassium (mg)')

# add title
plt.title('Honest Bar Chart')

plt.show()
```

With the adjusted y-axis limits, we can now see that the difference in both cereals' potassium content isn't as huge as we thought it was.
