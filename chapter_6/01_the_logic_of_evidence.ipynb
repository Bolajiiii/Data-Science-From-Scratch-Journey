# The Logic of Evidence

This notebook explores how probability changes when we condition on new information, and how Bayes’ Theorem formalizes this process of updating beliefs.

```python
# import libraries
import pandas as pd
```

```python
# load dataset
# !pip install palmerpenguins

from palmerpenguins import load_penguins
df = load_penguins()

```

This notebook will be focusing specifically on Conditional Probability. To start, we will be creating a [Contingency Table](https://www.geeksforgeeks.org/python/contingency-table-in-python/).

```python
data = pd.crosstab(index=df['species'],
                    columns=df['island'], margins=True)
```

A contigency table is a matrix used in statistics to display the frequency distribution of two or more categorical variables simultaneously. 

Now that we've created our contigency table, let's start asking questions.

**What is the probability that a selected penguin is a `Chinstrap` given that it's from `Dream` island?**


Before calculating the probability, let's form an expection.

If Dream island contains a large number of Chinstrap penguins, then the probability of a penguin being Chinstrap given that it is from Dream island should be relatively high.

```python
p_chinstrap_and_dream = data.loc['Chinstrap', 'Dream'] / data.loc['All', 'All']
p_dream = data.loc['All', 'Dream'] / data.loc['All', 'All']

conditional = round(p_chinstrap_and_dream / p_dream, 3)
print(f"The probability that a selected penguin is a Chinstrap given that it's from Dream island is {conditional}.")
```

Now let's move on to the big guy — **Bayes' Theorem**.

Bayes' Theorem is a mathematical formula used to calculate the likelihood of an event occurring based on prior knowledge of related conditions. It updates existing beliefs (prior probability) with new evidence to determine a revised probability (posterior probability).

We will be calculating the probability of a penguin being on `Dream` given that it's an `Adelie`.


$$P(D|A) = \frac{P(A|D) \cdot P(D)}{P(A)}$$

```python
# Prior: Probability of the penguin being on "Dream"
prior_dream = data.loc['All', 'Dream'] / data.loc['All', 'All']

# Likelihood: Probaility of the penguin being an "Adelie" given that it's on "Dream"
likelihood_adelie_dream = data.loc['Adelie', 'Dream'] / data.loc['All', 'Dream']

# Evidence: Probability of the penguin being an "Adelie"
evidence_adelie = data.loc['Adelie', 'All'] / data.loc['All', "All"]

# Posterior: Probability of the penguin being on "Dream" given that it's an "Adelie"
post_dream_adelie = (likelihood_adelie_dream * prior_dream) / evidence_adelie
print(f"The probability of a penguin being on Dream given that it's an Adelie is {post_dream_adelie}")
```

To confirm our result, we can compute this probability directly from the contingency table. 

```python
prob_dream_adelie = data.loc['Adelie', 'Dream'] / data.loc['Adelie', 'All']
print(f"The probability of a penguin being on Dream given that it's an Adelie is {prob_dream_adelie}")
```

By comparing unconditional and conditional probabilities, we saw that knowing additional context (such as the island a penguin belongs to) can significantly shift our expectations.

Bayes’ Theorem formalizes this idea by providing a structured way to update our beliefs based on evidence. Instead of treating probabilities as fixed, it allows us to refine them as new information becomes available.

This is especially important in real-world scenarios where decisions must be made under uncertainty. Bayes’ Theorem provides a logical framework for learning from evidence and making more informed predictions.
