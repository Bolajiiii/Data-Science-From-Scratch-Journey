# The Shape of Uncertainty

Random outcomes often feel messy and unpredictable. But when we observe them over time, patterns begin to emerge. This notebook explores how these patterns take shape, and how probability distributions help us make sense of uncertainty.

```python
# import libraries

import random
import numpy as np
import matplotlib.pyplot as plt
```

### Discrete Values


Rolling a die is random, and you can’t predict what you’ll get each time. But if you repeat the process many times, patterns start to appear in how often each outcome shows up.

To explore this, I will simulate rolling a die multiple times and record how often each outcome appears.

```python
# simulate a 6 sided die rolling 100 times
dice_rolls = [random.randint(1, 6) for _ in range(100)]
# visualize dice rolls
plt.hist(dice_rolls, bins=np.arange(1, 8) - 0.5, rwidth=0.9)
plt.xticks(range(1, 7))
plt.title("Simulation of 100 Dice Rolls")
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.show()
```

With a smaller number of rolls, the results are quite uneven. Some numbers appear more frequently than others, and it doesn’t look very balanced.

**What happens if we increase the number of rolls?**

```python
# simulate a 6 sided die rolling 10,000 times
dice_rolls = [random.randint(1, 6) for _ in range(10_000)]
# visualize dice rolls
plt.hist(dice_rolls, bins=np.arange(1, 8) - 0.5, rwidth=0.9)
plt.xticks(range(1, 7))
plt.title("Simulation of 10,000 Dice Rolls")
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.show()
```

At first, the results seem very random, and it’s hard to see any pattern. But as the number of dice rolls increases, the outcomes start to even out. They are still random, but consistent enough to reveal the underlying probabilities.

### Continuous Values


So far, the outcomes we’ve looked at fall into fixed categories, like the faces of a die. But not all random processes work like that.

Sometimes, values can be anything within a range, instead of a set of distinct outcomes. That means we can’t really count them in the same way anymore.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 100 data points
continuous_data_100 = np.random.uniform(0, 1, 100)
axes[0].hist(continuous_data_100, bins=20)
axes[0].set_title("Histogram of 100 Continuous Random Values")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

# 1000 data points
continuous_data_1000 = np.random.uniform(0, 1, 1000)
axes[1].hist(continuous_data_1000, bins=20)
axes[1].set_title("Histogram of 1,000 Continuous Random Values")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")

# 10,000 data points
continuous_data_10000 = np.random.uniform(0, 1, 10_000)
axes[2].hist(continuous_data_10000, bins=20)
axes[2].set_title("Histogram of 10,000 Continuous Random Values")
axes[2].set_xlabel("Value")
axes[2].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

With fewer samples, the distribution looks uneven, mostly due to random variation. But as the number of values increases, it starts to look more uniform. Instead of focusing on individual values, you begin to notice how the data is spread across the range.

Just like the dice rolls became more balanced over time, the continuous data also starts to settle into a clearer shape. This shows that uncertainty isn’t only about counts, it can also be seen in how the values are distributed.
