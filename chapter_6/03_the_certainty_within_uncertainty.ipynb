# The Certainty Within Uncertainty


This notebook explores the magic of the Central Limit Theorem, which shows that the average of many random variables tends to follow a normal distribution, even if the original data is not normally distributed. This provides the certainty we need to make predictions in an uncertain world.

```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
```

### Uniform Distribution

We’ll start with a uniform distribution, where all values within a range are equally likely. At first, this type of data looks flat and unstructured. But as we begin to work with samples and averages, something interesting starts to happen.

```python
# Generate uniform distribution
uniform_dist = np.random.uniform(0, 1, 30_000)

# Plot distribution
plt.hist(uniform_dist, bins=30)
plt.title('Uniform Distribution with 30,000 Values')
plt.show()
```

We can observe that the distribution has a flat (or almost flat) shape. But the magic of CLT will turn this into a normal distribution.

```python
# Take sample from distribution
s = 30
sample_means = [np.mean(np.random.choice(uniform_dist, size=s)) for _ in range(1000)]

# Plot sample means 
plt.hist(sample_means, bins=30, edgecolor='white')
plt.title(f"Distribution of 1000 Sample Means (n={s})")
plt.show()
```

The distribution of the sample means now takes on a clear bell shape, even though the original data was uniform.

### Skewed Distribution

What if we were working with a skewed distribution. It sounds unbelievable that some averages from a skewed distribution can form a normal distrubution. Let's confirm this!

```python
# Generate exponential distribution
exponential_dist = np.random.exponential(scale=1.0, size=30_000)

# Plot distribution
plt.hist(exponential_dist)
plt.title("Exponential Distribution with 30,000 values")
plt.show()
```

This distribution appears to be very skewed. Let's see the magic of the CLT.

```python
# Take sample from distribution
s = 30
sample_means = [np.mean(np.random.choice(exponential_dist, size=s)) for _ in range(1000)]

# Plot sample means 
plt.hist(sample_means, bins=30, edgecolor='white')
plt.title(f"Distribution of 1000 Sample Means (n={s})")
plt.show()
```

Another normal distribution! We have now the seen the effect of the Central Limit Theorem.

This means that even when individual data points are unpredictable, the behavior of their averages becomes stable and predictable.
