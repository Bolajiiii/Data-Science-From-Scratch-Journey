# Matrix Operations

```python
# import libraries
import pandas as pd
import numpy as np
```

**What are Matrices?**

A matrix is a two-dimensional array of numbers arranged in rows and columns.

Matrices are used to represent datasets, where each row corresponds to an observation or sample, and each column represents a feature or attribute of that sample.

In this project, I'll be using a movie dataset where each row represents a movie, and each column represents a feature.


```python
# load data into dataframe
df = pd.read_csv('../datasets/Hydra-Movie-Scrape.csv')
```

```python
# create movie feature matrix
movie_matrix = df[['Runtime', 'Rating']].to_numpy()
```

Now that we have our feature matrix, let's check its shape

```python
movie_matrix.shape
```

It is a 3886 x 2 matrix, meaning it has 3886 rows and 2 columns. Since the number of rows is much larger than the number of columns, this is a tall matrix.

```python
# extract movie titles
movie_titles = df['Title']
```

Let us check out the rows and columns.

```python
# calculate mean of rows
np.mean(movie_matrix[:5], axis=1) # first 5 rows
```

```python
# calculate mean of columns
np.mean(movie_matrix, axis=0)
```

The mean runtime for this dataset is 100.44 minutes and mean rating is 6.56

```python
# find max runtime
runtime_col = movie_matrix[:, 0]
max_run = float(np.max(runtime_col))
id_run = np.argmax(runtime_col)

print(f"The movie with the longest runtime is \"{movie_titles[id_run]}\" with a runtime of {round(max_run/60, 2)} hours.")
```

A 5 hour 6 minutes long movie? Damn, I wonder what kind of story needs that much time to tell.

```python
# find max rating
rating_col = movie_matrix[:, 1]
max_rate = float(np.max(rating_col))
id_rate = np.argmax(rating_col)

print(f"The movie with the highest rating is \"{movie_titles[id_rate]}\" with a rating of {max_rate}")
```

Now let's check out the Transpose of the matrix.

```python
movie_transpose = movie_matrix.T
```

```python
print(f"First five rows of original matrix: \n{movie_matrix[:5]}\n")

print(f"First five columns of transposed matrix: \n{movie_transpose[:, :5]}")
```

In this notebook, the movie dataset was represented as a matrix where rows correspond to movies and columns represent numerical features. Viewing the data in matrix form makes it possible to apply linear algebra operations to summarize and transform the dataset.
