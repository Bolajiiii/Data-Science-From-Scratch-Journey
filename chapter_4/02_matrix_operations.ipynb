# Matrix Operations

```python
# import libraries
import pandas as pd
import numpy as np
```

**What are Matrices?**

In simple terms, matrix is a two-dimensional array of numbers arranged in rows and columns.

Matrices are used to represent datasets, where each row corresponds to an observation or sample, and each column represents a feature or attribute of that sample.

In this project, I'll be using a movie dataset where each row represents a movie, and each column represents a feature.


```python
# load data into dataframe
df = pd.read_csv('../datasets/Hydra-Movie-Scrape.csv')
```

```python
# create movie feature matrix
movie_matrix = df[['Runtime', 'Rating']].to_numpy()
```

Now that we have our feature matrix, let's check its shape

```python
movie_matrix.shape
```

It is a 3886 x 2 matrix, meaning it has 3886 rows and 2 columns. Our matrix is tall and skinny.

```python
# extract movie titles
movie_titles = df['Title']
```

Let us check out the rows and columns.

```python
# calculate mean of rows
np.mean(movie_matrix, axis=1)
```

```python
# calculate mean of columns
np.mean(movie_matrix, axis=0)
```

The mean runtime for this dataset is 100.44 minutes and mean rating is 6.56

```python
# calculate max runtime
runtime_col = movie_matrix[:, 0]
max_value = float(max(runtime_col))
id_max = np.argmax(runtime_col)

print(f"The movie with the longest runtime is \"{movie_titles[id_max]}\" with a runtime of {round(max_value/60, 2)} hours.")
```
